```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import decomposition
from sklearn.decomposition import PCA, TruncatedSVD
```

```python
All = pd.read_csv('df_ALL.csv')
All.head()
```

```python
target = All["Cancer"]
target_names = ["ALL", "AML"]
```

```python
data = All.drop("Cancer", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=40)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
#SVM without SVD
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
#KNN without SVD
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train, y_train)
print('k=13 Test Acc: %.3f' % knn.score(X_test, y_test))
```

```python
#DecTree without SVD
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#RandForest without SVD
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
f = sorted(zip(rf.feature_importances_, feature_names), reverse=True)
def feature_names_Number(n):
    sel=[t[1] for t in f[:n]]
    return sel
```

```python
data = All.drop("Cancer", axis=1)
n = 50
feature_names = data[feature_names_Number(n)]
data.head()
```

```python
#Implementing the SVD than trying our models again.
svd = TruncatedSVD(n_components=10)
x = svd.fit(feature_names).transform(feature_names)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, target, random_state=41)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
#SVM with SVD
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
#KNN with SVD
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train, y_train)
print('k=13 Test Acc: %.3f' % knn.score(X_test, y_test))
```
